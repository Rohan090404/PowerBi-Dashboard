```python
def replace_outliers (x):
    numerical_cols = x.select_dtypes(["int","float"]).columns
    print(numerical_cols)
    for i in numerical_cols:
        q1=x[i].quantile(0.25)
        q3=x[i].quantile(0.75) 
        iqr=q3-q1
        lw=q1-1.5*iqr 
        uw=q3+1.5*iqr 
        print(f"{i}:lw:{lw},uw:{uw}")
        x[i]=x[i].apply(lambda x:min(max(x,lw),uw))    # lambda  min(max(79545,40026),97237)
    return x
df=replace_outliers(df)
```

```python
'''
apply() : This method can be applied to both series or data frames. on series it applies the function element wise. on a data frame it applues the function
along an axis that is rows and column and can return as series or scaler(single value) 

map() : it is used to substitute each value in a series with another value ,it takes a dictionary or a function as an argument when a dictionary is provided
it maps values based on the keys and when a function is provided it applies the function to each element of a series
'''
```

```python
features=df.drop("Price",axis=1)
target=df["Price"]
```

```python
features.head(2)
```

```python
target.head()
```

```python
# step 4 Data scaling
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
features.iloc[:]=scaler.fit_transform(features.iloc[:])
features.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    features, target, test_size=0.2, random_state=50)  # 20% in test and 80% in train
```

```python
'''
scikit-learn or sklearn is an open source python library used for machine learning it consist of ready to use tools for various machine learning task
,here instead of building complex algorithm from sractch you can use scikit learning to easily implement them it helps in doing classification ,regression,
clustering,data preprocessing,model selection and evaluation,it is build on other fundamental python libraries like numpy (for numerical operation on multi-
dimensional arrays) and scipy for scientific computing it works well with matplotlib for data visual and its consistent interface makes it relatively easy
to learn and use for beginers
'''
```

```python
print(xtrain.shape,ytrain.shape)
```

```python
print(xtest.shape,ytest.shape)
```

```python
from sklearn.linear_model import LinearRegression  # sklearn : library , linear_model : module
reg= LinearRegression()   # reg is object of class LinearRegression
reg.fit(xtrain,ytrain)     # training, here fit find best value of m and c 

ypred=reg.predict(xtest)  # training done ! , so now prediction is doing it is puting x in equation and giving y that is output
                          # y=mx+c
```

```python
reg.coef_,
```

```python
f=pd.DataFrame(
    {
        "columns":features.columns,
        "coef":reg.coef_,
        "intercept" : reg.intercept_
    }
)
f
```

```python
reg.intercept_
```

```python
reg.score(xtrain,ytrain)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
mae=mean_absolute_error(ytest,ypred)
mse=mean_squared_error(ytest,ypred)
rmse=np.sqrt(mse)
r2=r2_score(ytest,ypred)
```

```python
print(f"MAE : {mae}\nMSE {mse}\nRMSE : {rmse}\nR2-Score :{r2}")
```

```python
# Homoscedasticity ( assumption 3 )  : it should not be funnel shape so it is satisfied
sns.scatterplot(x=ypred,y=(ytest-ypred))
```

```python
# Normality ( assumption 4) satisfied
sns.kdeplot(ytest-ypred)
```
